```python
df_classes = df_classes.drop(["Unnamed: 0", "_id", "index"],  axis=1)
```

```python
import ast
for class_name, proficience, proficiency_choice, saving_throws, starting_equipment, class_level, subclass, spells in zip(df_classes["name"], df_classes["proficiencies"], df_classes["proficiency_choices"], df_classes["saving_throws"], df_classes["starting_equipment"], df_classes["class_levels"], df_classes["subclasses"], df_classes["spellcasting"]):
    print(f"For {class_name}")
    proficience = ast.literal_eval(proficience)
    print(f"Proficience: {", ".join([str(i['name']) for i in proficience])}")
    proficiency_choice = ast.literal_eval(proficiency_choice)
    #print(f"Proficiency Choices: \n{"\n".join([str(pc) for pc in proficiency_choice])}")
    #print(len(proficiency_choice))
    #print(saving_throws)
    saving_throws = ast.literal_eval(saving_throws)
    print(f"Saving Throws: {', '.join([str(st['name']) for st in saving_throws])}")

    # print(starting_equipment) --- оказалось бесполезным, там url и название класса
    # print(class_level) --- и это тоже

    subclass = ast.literal_eval(subclass)
    print(f"Subclasses: {', '.join([str(sb['name']) for sb in subclass])}")
    if pd.isna(spells):
        print("Have not spells")
    else:
        print("Have spells")
    print("\n")
    i = 1
    for pc in proficiency_choice:
        print(f"Proficiency Choice {i}:")
        print(f"choose: {pc['choose']}")
        print(f"from: {", ".join([str(fr["name"]) for fr in pc['from']])}")
        print("\n")
        i += 1
    print("===========")

```
